# Combining Data

```python
expense_only.info()
```

```python
expense_only_2.info()
```

```python
expense_only_2.rename(columns={'category': 'Category', 'amount': 'Amount', 'title':'Transaction Description'},inplace=True)
expense_only_2['Amount'] = pd.to_numeric(expense_only_2['Amount'], errors='coerce')
```

```python
combined_df = pd.concat([expense_dataset, expense_only_2], ignore_index=True)
```

```python
combined_df.info()
```

```python
combined_df = combined_df[['Date','Transaction Description','Category','Amount']]
```

```python
combined_df['Amount']
```

```python
combined_df.isna().sum()
```

```python
combined_df.Category.unique()
```

```python
combined_df['Category'].replace({'Food & Drink':'Food','Food & Drinks':'Food','Entertainment':'Other','Transport':'Travel',
                                 'Bills & Fees':'Utilities','Health & Fitness':'Health'},inplace=True)
```

```python
combined_df.Category.unique()
```

```python
combined_df.dropna(inplace=True)
non_salary_condition = combined_df['Category'] != 'Salary'
combined_df = combined_df[non_salary_condition]
```

```python
combined_df.info()
```

```python
dff = (combined_df['Category']=='Salary')
```

```python
# Use the boolean mask (dff) to filter the original combined_df
salary_entries = combined_df[dff]

# Display the resulting DataFrame containing only 'Salary' rows
print(salary_entries)
```

```python
combined_df.isna().sum()
```
